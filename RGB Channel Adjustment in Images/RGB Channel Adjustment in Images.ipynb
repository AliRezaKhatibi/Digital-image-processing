```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def adjust_channel(image, r_factor=1.0, g_factor=1.0, b_factor=1.0):
    # Split color channels
    b, g, r = cv2.split(image)
    
    # Apply intensity factor to each channel
    b = np.clip(b * b_factor, 0, 255).astype(np.uint8)
    g = np.clip(g * g_factor, 0, 255).astype(np.uint8)
    r = np.clip(r * r_factor, 0, 255).astype(np.uint8)
    
    # Merge channels back
    adjusted_image = cv2.merge([b, g, r])
    return adjusted_image

# Read the image
image_path = 'data/image (1).jpg'  # Change the path if needed
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Set channel intensity factors
r_factor = 0.5  # Red intensity
g_factor = 1  # Green intensity
b_factor = 0.5  # Blue intensity

adjusted_image = adjust_channel(image, r_factor, g_factor, b_factor)

# Show images in Jupyter Notebook
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")
ax[1].imshow(adjusted_image)
ax[1].set_title("Adjusted Image")
ax[1].axis("off")
plt.show()

# Save the adjusted image
output_dir = "output"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "adjusted_image.jpg")
cv2.imwrite(output_path, cv2.cvtColor(adjusted_image, cv2.COLOR_RGB2BGR))
print(f"Adjusted image saved at: {output_path}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Load the image
image = cv2.imread(r'data\image (11).jpg')  # Replace 'input_image.jpg' with your image path

# Convert the image to RGB (OpenCV loads images in BGR format by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Iterate over all pixels and replace red intensity with blue
# We only replace red pixels with significant intensity (Red > 100 for example)
for i in range(image_rgb.shape[0]):
    for j in range(image_rgb.shape[1]):
        # Check if the red intensity is higher than a threshold
        if image_rgb[i, j, 0] > 100:  # Red channel
            # Copy the red intensity to the blue channel
            image_rgb[i, j, 2] = image_rgb[i, j, 0]
            image_rgb[i, j, 0] = 0  # Set the red channel to 0 (remove red)

# Define the output folder and create it if it doesn't exist
output_folder = 'processed_images'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Save the final image
output_path = os.path.join(output_folder, 'output_image_red_to_blue_intensity.jpg')
cv2.imwrite(output_path, cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR))  # Save as BGR

# Display the original and processed image side by side
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')  # Hide axes

# Processed Image
plt.subplot(1, 2, 2)
plt.imshow(image_rgb)
plt.title('Processed Image (Red to Blue based on Intensity)')
plt.axis('off')  # Hide axes

plt.show()

# Print output path to confirm saving
print(f"Image saved at: {output_path}")
```
